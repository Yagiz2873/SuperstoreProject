### Capping yöntemi

```python
cap_df = df.copy()

cap_df['Sales'] = np.where(
    (cap_df['Segment'] == 'Consumer') & (cap_df['Sales'] >= upper_limit),
    upper_limit,
    np.where(
        (cap_df['Segment'] == 'Consumer') & (cap_df['Sales'] <= lower_limit),
        lower_limit,
        cap_df['Sales']
    )
)
```

```python
def remove_outliers(df,col):
    Q1 = df[col].quantile(0.05)
    Q3 = df[col].quantile(0.95)
    IQR = Q3-Q1

    filter = (df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)
    removed_count = np.sum(filter==False)
    tot = len(df)
    df = df.loc[filter]
    print(f"Removed {removed_count} outliers from {tot} rows.")
    return df
```

```python
df = remove_outliers(df, 'Sales')
```

```python
df = remove_outliers(df, 'Quantity')
```

```python
df = remove_outliers(df, 'Profit')
```
